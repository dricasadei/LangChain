# Considerações:
- Exemplo usando LangChain utilizando Chains de forma simplificada

## Tipos de exemplos:


# Setup

```python
# importando as bibliotecas

from langchain_openai import OpenAI
from langchain.prompts import PromptTemplate
import yaml
import os
```

```python
# lendo o arquivo config.yaml para ter acesso a chave de acesso da OpenAI

with open('config.yaml', 'r') as config_file:
    config = yaml.safe_load(config_file)
os.environ['OPENAI_API_KEY'] = config['OPENAI_API_KEY']
```

```python
openai = OpenAI(model_name='gpt-3.5-turbo-instruct', temperature=0) # instanciando um objeto OpenAI
```

# Simple

```python
prompt_template = PromptTemplate.from_template("Descreva as tendências tecnológicas em {ano}.") # instanciando o objeto do PromptTemplate
runnable_sequence = prompt_template | openai # encadeamento dos objetos prompt_template e openai
output = runnable_sequence.invoke({"ano": "2024"}) # chamando o metodo invoke para executar a sequencia indicada na variável runnable_sequence
```

```python
print("Output:\n", output) # imprimindo a saida
```
