# Considerações:
- Exemplo usando LangChain utilizando Tools
- [LangChain Tools](https://python.langchain.com/v0.1/docs/modules/tools/)

# Setup

```python
# importando as bibliotecas

from langchain_experimental.utilities import PythonREPL
from langchain_community.tools import DuckDuckGoSearchRun,WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

# DuckDuckGoSearchRun: ferramenta de busca sem ser necessário chave de acesso de alguma api, como do google
# WikipediaQueryRun: ferramenta da wikipedia.
# WikipediaAPIWrapper: ferramenta para fazer a consulta na wikipedia
# perceba que as tools em si, não utilizam os modelos, no próximo exemplo será mostrado um exempli mais completo
```

# Executa Código Python

```python
python_repl = PythonREPL() 
result = python_repl.run("print(5 * 5)")
print(result)

# PythonREPL(): dependendo do que se pede aparece mensagem de warning, como abaixo. 
```

# Busca na Web

```python
ddg_search = DuckDuckGoSearchRun() # instância do objeto DuckDuckGoSearchRun()

query = "Qual é a capital da Guiana?" # consulta
# search_results = ddg_search.run(query) # executa a busca

try:
    # Executa a busca
    search_results = ddg_search.run(query)
    print(search_results)
except Exception as e:
    print(f"Erro ao executar busca: {e}")

# como é uma api pública tem um limite de acesso, por isso, ao rodar, pode aparecer erro: RatelimitException
```

# Consulta a Wikipedia

```python
wikipedia = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper()) # instância do objeto WikipediaQueryRun
query = "Guiana" # termo da busca/consulta
wikipedia_results = wikipedia.run(query)
```

```python
print("Resultados da Wikipedia:", wikipedia_results)
```

- São exemplos isolados de tools, mas em geral, são usadas de forma integrada como com os agents.
